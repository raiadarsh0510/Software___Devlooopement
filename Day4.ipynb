***ASYMPTOTIC NOTATION***

Asymptotic notation is a mathematical way to describe how the running time or memory usage of an algorithm grows as the input size (n) becomes very large. It helps compare algorithms without focusing on specific hardware or implementation details.

*The three most common asymptotic notations are:*

**Big O Notation (O)** –

Upper Bound (Worst Case).

Describes the maximum growth rate of an algorithm.
Example:

Linear search: O(n)

Binary search: O(log n)

Bubble sort (worst case): O(n²)

**Big Omega Notation (Ω)** –

Lower Bound (Best Case)

Describes the minimum growth rate of an algorithm.

Example:

Linear search (element is first): Ω(1)

**Big Theta Notation (Θ**) –

Tight Bound (Average/Exact Growth Rate)

Describes when the upper and lower bounds are the same.

Example:

Accessing an array element: Θ(1)
Merge sort: Θ(n log n)


***Why Asymptotic Notation Is Important***

Compares algorithms efficiently.
Predicts performance for large inputs.
Ignores constants and lower-order terms.
Helps choose the most efficient algorithm.

Example:

If an algorithm takes 3n² + 5n + 10 operations, its asymptotic complexity is:

O(n²)

because as n becomes very large, the n² term dominates while constants and lower-order terms become insignificant.
