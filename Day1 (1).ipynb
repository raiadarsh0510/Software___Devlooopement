```python
# Restaurant Menu Project

menu = {
    "coffee": 80,
    "cold coffee": 120,
    "pizza": 250,
    "burger": 150,
    "sandwich": 100,
    "tea": 40,
    "cake": 90
}


print("     WELCOME TO SURYA CAFE")
print("   Fresh Food, Happy Mood!")


print("\n----- MENU -----")
for item, price in menu.items():
    print(f"{item.title()} : ₹{price}")

order = input("\nEnter the food item: ").lower()

if order in menu:
    quantity = int(input("Enter quantity: "))

    price = menu[order]
    total = price * quantity

    order_number = 1001


    print("          ORDER RECEIPT")

    print("Order Number :", order_number)
    print("Item         :", order.title())
    print("Price        : ₹", price)


    print("Total Bill   : ₹", total)

else:
    print("\nSorry! This item is not available in our menu.")
```
