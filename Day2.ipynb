```python
# List of names
l1 = ["Aman", "Rahul", "Priya", "Sneha", "Rohan", "Anjali", "Karan"]

print("Printing names using for loop, if, continue, and break:\n")

for name in l1:
    # Skip Rahul
    if name == "Rahul":
        print(name, "- skipped")
        continue

    # Stop the loop when Karan is found
    if name == "Karan":
        print(name, "- found, loop stopped")
        break

    # Print other names
    print(name)
```

```python
# Number list
numbers = [10, 20, 30, 40, 50, 60]

# 1. Print all elements using range()
print("All elements:")
for i in range(len(numbers)):
    print(numbers[i])

# 2. Print index and element
print("\nIndex and Element:")
for i in range(len(numbers)):
    print("Index", i, "=", numbers[i])

# 3. Print first 3 elements
print("\nFirst 3 elements:")
for i in range(3):
    print(numbers[i])

# 4. Print elements from index 2 to 4
print("\nElements from index 2 to 4:")
for i in range(2, 5):
    print(numbers[i])

# 5. Print alternate elements (step = 2)
print("\nAlternate elements:")
for i in range(0, len(numbers), 2):
    print(numbers[i])

# 6. Print elements in reverse order
print("\nReverse order:")
for i in range(len(numbers) - 1, -1, -1):
    print(numbers[i])
```

```python
# Original set
s = {10, 20, 30, "Aman", "Rahul", "Priya"}

print("Original Set:", s)

# 1. Add one element
s.add(40)
print("After add():", s)

# 2. Update the set with multiple elements
s.update([50, "Rohan"])
print("After update():", s)

# 3. Remove an element
s.remove(20)
print("After remove():", s)

# 4. Discard an element
s.discard("Priya")
print("After discard():", s)

# 5. Pop a random element
removed = s.pop()
print("Removed element:", removed)
print("After pop():", s)

# 6. Length of the set
print("Length of set:", len(s))

# 7. Membership test
if "Aman" in s:
    print("Aman is present")
else:
    print("Aman is not present")

# 8. Copy the set
s2 = s.copy()
print("Copied set:", s2)

# 9. Union
a = {1, 2, 3}
b = {3, 4, 5}
print("Union:", a.union(b))

# 10. Intersection
print("Intersection:", a.intersection(b))

# 11. Difference
print("Difference:", a.difference(b))

# 12. Symmetric Difference
print("Symmetric Difference:", a.symmetric_difference(b))

# 13. Clear the set
s2.clear()
print("After clear():", s2)
```

```python
# Dictionary
student = {
    "name": "Aman",
    "age": 20,
    "marks": [85, 90, 95],
    "subjects": ("Python", "C", "Java"),
    "skills": {"Coding", "Typing"},
    "address": {
        "city": "Lucknow",
        "state": "Uttar Pradesh"
    }
}

# 1. Print dictionary
print("Original Dictionary:")
print(student)

# 2. Access values
print("\nName:", student["name"])
print("Age:", student["age"])
print("Marks:", student["marks"])
print("City:", student["address"]["city"])

# 3. Add a new key-value pair
student["course"] = "BCA"
print("\nAfter adding course:")
print(student)

# 4. Update a value
student["age"] = 21
print("\nAfter updating age:")
print(student)

# 5. Update multiple values
student.update({"semester": 3, "college": "ABC College"})
print("\nAfter update():")
print(student)

# 6. Delete a key
del student["course"]
print("\nAfter del:")
print(student)

# 7. Remove using pop()
student.pop("semester")
print("\nAfter pop():")
print(student)

# 8. Print all keys
print("\nKeys:")
print(student.keys())

# 9. Print all values
print("\nValues:")
print(student.values())

# 10. Print all key-value pairs
print("\nItems:")
print(student.items())

# 11. Check if a key exists
if "name" in student:
    print("\n'name' key is present")

# 12. Length of dictionary
print("\nLength:", len(student))

# 13. Copy dictionary
student2 = student.copy()
print("\nCopied Dictionary:")
print(student2)

# 14. Loop through dictionary
print("\nLooping through dictionary:")
for key, value in student.items():
    print(key, ":", value)

# 15. Clear copied dictionary
student2.clear()
print("\nAfter clear():")
print(student2)
```
