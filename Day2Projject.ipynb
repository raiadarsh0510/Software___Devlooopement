```python
entry = int(input("Number of Students? "))

data = []

for i in range(entry):
    print(f"\nEntering data for Student {i+1}")

    name = input("Enter name: ")

    subjects = {}
    total_marks = 0

    sub_count = int(input("How many subjects? "))

    for j in range(sub_count):
        sub_name = input(f"Enter subject {j+1} name: ")
        mark = float(input(f"Enter marks for {sub_name}: "))
        subjects[sub_name] = mark
        total_marks += mark

    average = total_marks / sub_count

    # Grade system
    if average >= 90:
        grade = "A+"
    elif average >= 75:
        grade = "A"
    elif average >= 60:
        grade = "B"
    elif average >= 40:
        grade = "C"
    else:
        grade = "Fail"

    data.append({
        "Name": name,
        "Subjects": subjects,
        "Total": total_marks,
        "Average": average,
        "Grade": grade
    })

# FINAL MARKSHEET OUTPUT
print("\n" + "="*40)
print("FINAL MARKSHEET")
print("="*40)
print("\n" + "="*40)
print("SRMCEM")
print("="*40)

for student in data:
    print("\nName:", student["Name"])
    print("Subjects:")
    for sub, mark in student["Subjects"].items():
        print(f"  {sub}: {mark}")

    print("Total Marks:", student["Total"])
    print("Average:", student["Average"])
    print("Grade:", student["Grade"])
    print("-"*40)
```

```python
entry = int(input("Number of Students? "))

# Take common subjects (ONLY ONCE)
sub_count = int(input("How many subjects (common for all)? "))

subjects_list = []

for i in range(sub_count):
    sub_name = input(f"Enter subject {i+1} name: ")
    subjects_list.append(sub_name)

data = []

# Student data entry
for i in range(entry):
    print(f"\nEntering data for Student {i+1}")

    name = input("Enter name: ")

    marks = {}
    total_marks = 0

    for sub in subjects_list:
        mark = float(input(f"Enter marks for {sub}: "))
        marks[sub] = mark
        total_marks += mark

    average = total_marks / sub_count

    # Grade system
    if average >= 90:
        grade = "A+"
    elif average >= 75:
        grade = "A"
    elif average >= 60:
        grade = "B"
    elif average >= 40:
        grade = "C"
    else:
        grade = "Fail"

    data.append({
        "Name": name,
        "Marks": marks,
        "Total": total_marks,
        "Average": average,
        "Grade": grade
    })

# FINAL MARKSHEET OUTPUT
print("\n" + "="*50)
print("SRMCEM FINAL MARKSHEET")
print("="*50)

for student in data:
    print("\nName:", student["Name"])
    print("Marks:")

    for sub, mark in student["Marks"].items():
        print(f"  {sub}: {mark}")

    print("Total Marks:", student["Total"])
    print("Average:", student["Average"])
    print("Grade:", student["Grade"])
    print("-"*50)
```

```python
entry = int(input("Number of Students? "))

sub_count = int(input("How many subjects (common for all)? "))

subjects_list = []

for i in range(sub_count):
    sub_name = input(f"Enter subject {i+1} name: ")
    subjects_list.append(sub_name)

data = []

for i in range(entry):
    print(f"\nEntering data for Student {i+1}")

    name = input("Enter name: ")

    marks = {}
    total_marks = 0

    for sub in subjects_list:
        mark = float(input(f"Enter marks for {sub}: "))
        marks[sub] = mark
        total_marks += mark

    average = total_marks / sub_count

    # Grade system
    if average >= 90:
        grade = "A+"
    elif average >= 75:
        grade = "A"
    elif average >= 60:
        grade = "B"
    elif average >= 40:
        grade = "C"
    else:
        grade = "Fail"

    data.append({
        "Name": name,
        "Marks": marks,
        "Total": total_marks,
        "Average": average,
        "Grade": grade
    })

# ================= TABLE OUTPUT =================
print("\n" + "="*80)
print("SRMCEM FINAL MARKSHEET".center(80))
print("="*80)

# Header
header = f"{'Name':<15}"

for sub in subjects_list:
    header += f"{sub:<10}"

header += f"{'Total':<10}{'Avg':<10}{'Grade':<10}"

print(header)
print("-"*80)

# Data rows
for student in data:
    row = f"{student['Name']:<15}"

    for sub in subjects_list:
        row += f"{student['Marks'][sub]:<10}"

    row += f"{student['Total']:<10.2f}{student['Average']:<10.2f}{student['Grade']:<10}"

    print(row)

print("="*80)7
```
