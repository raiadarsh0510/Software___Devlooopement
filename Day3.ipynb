***OOPS***
What is OOP (Object-Oriented Programming)?

OOP is a programming style based on objects and classes.

A class is a blueprint, and an object is its real instance.

Why do we use OOP?

We use OOP because it makes programs:

1. Easier to manage (Modularity)

Code is divided into small parts (classes).

2. Reusable (Reusability)

You can reuse classes in different programs.

3. Secure (Encapsulation)

Data is protected inside objects.

4. Flexible (Polymorphism)

Same function name can work differently.

5. Easy to maintain

Changes in one part do not affect the whole program.

***EXPENSE TRACKER***

```python
class ExpenseTracker:
    def __init__(adarsh):
        adarsh.expenses = []

    def add_expense(adarsh, amount, category):
        adarsh.expenses.append({"amount": amount, "category": category})

    def get_expenses_by_category(adarsh, category):
        expenses_in_category = [expense for expense in adarsh.expenses if expense["category"] == category]
        return expenses_in_category

    def get_total_expenses(adarsh):
        total_expenses = sum(expense["amount"] for expense in adarsh.expenses)
        return total_expenses



```

```python
class ExpenseTracker:
    def __init__(self):
        self.expenses = []

    # Add a new expense
    def add_expense(self, amount, category, description):
        self.expenses.append({
            "amount": amount,
            "category": category,
            "description": description
        })

    # Get expenses by category
    def get_expenses_by_category(self, category):
        return [expense for expense in self.expenses
                if expense["category"].lower() == category.lower()]

    # Calculate total expenses
    def get_total_expenses(self):
        return sum(expense["amount"] for expense in self.expenses)

    # Display all expenses
    def display_expenses(self):
        if not self.expenses:
            print("\nNo expenses recorded.")
            return

        print("\n------ All Expenses ------")
        for i, expense in enumerate(self.expenses, start=1):
            print(f"{i}. {expense['description']} | "
                  f"Category: {expense['category']} | "
                  f"Amount: ₹{expense['amount']}")
        print("--------------------------")


# Create ExpenseTracker object
tracker = ExpenseTracker()

# Menu
while True:
    print("\n====== Expense Tracker ======")
    print("1. Add Expense")
    print("2. View All Expenses")
    print("3. Search by Category")
    print("4. Show Total Expenses")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        description = input("Enter Description: ")
        amount = float(input("Enter Amount: ₹"))
        category = input("Enter Category (Food/Shopping/Bills/Travel/etc.): ")

        tracker.add_expense(amount, category, description)
        print("Expense added successfully!")

    elif choice == "2":
        tracker.display_expenses()

    elif choice == "3":
        category = input("Enter Category: ")
        expenses = tracker.get_expenses_by_category(category)

        if expenses:
            print(f"\nExpenses in '{category}' category:")
            for expense in expenses:
                print(f"{expense['description']} - ₹{expense['amount']}")
        else:
            print("No expenses found in this category.")

    elif choice == "4":
        print(f"\nTotal Expenses: ₹{tracker.get_total_expenses():.2f}")

    elif choice == "5":
        print("Thank you for using Expense Tracker!")
        break

    else:
        print("Invalid choice! Please try again.")
```
